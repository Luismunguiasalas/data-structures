```python
# convert max heap to min heap

def convert_max_heap(arr):
    def percolate_up(arr, index):
        parent = index // 2

        if index < 0:return

        if arr[parent] > arr[index]:
            arr[parent], arr[index] = arr[index], arr[parent]
            percolate_up(arr, parent)
    
    for i in range(len(arr)):
        percolate_up(arr, i)

    ans = []
    for n in arr:
        ans.append(n)

    return ans


def convert_max_heap(arr):
    def heapify(arr, index):
        left = (index * 2) + 1
        right = (index * 2) + 2
        small = index

        if len(arr) > left and arr[left] < arr[small]:
            small = left

        if len(arr) > right and arr[right] < arr[small]:
            small = right

        if small != index:
            arr[index], arr[small] = arr[small], arr[index]
            heapify(arr, small)
    for i in range(len(arr)-1,-1,-1):
        heapify(arr,i)
    return arr
```

```python
# find k smallest elements in array

def find_k_smallest(lst,k):
    heap = MinHeap()
    heap.buildHeap(lst)
    ans = [heap.removeMin() for i in range(k)]
    return ans
```
